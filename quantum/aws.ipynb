```python
from subprocess import run
import boto3

env = run("aws-vault exec josue-pucp -- env".split(" "), capture_output=True)
env = env.stdout.decode("utf-8").split("\n")
env = dict([i.split("=", 1) for i in env if i.startswith("AWS")])
session = boto3.Session(
	region_name="us-east-1",
    aws_access_key_id=env["AWS_ACCESS_KEY_ID"],
    aws_secret_access_key=env["AWS_SECRET_ACCESS_KEY"],
    aws_session_token=env["AWS_SECURITY_TOKEN"]
)
```

```python
# import the QPU module
from braket.aws import AwsSession, AwsDevice
# Initialize Braket session with Boto3 Session credentials
aws_session = AwsSession(boto_session=session)
# choose the Rigetti device
device = AwsDevice("arn:aws:braket:us-west-1::device/qpu/rigetti/Aspen-M-3", aws_session)

# take a look at the device connectivity graph
device.properties.dict()['paradigm']['connectivity']
```
